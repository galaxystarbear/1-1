# 아래 예제코드를 참고해서 자신만의 자동 미분 문제를 만들어서 실행하고 결과를 분석하시오.

```python
from torch import FloatTensor
from torch.autograd import Variable
```

```python
a = Variable(FloatTensor([4]))
print(a)
```

```python
weights = [Variable(FloatTensor([i]), requires_grad=True) for i in (2, 5, 9, 7)]
```

```python
# unpacking (언패킹)을 통해 weights가 포함하는 값들을 서로 다른 tensor 변수에 저장
w1, w2, w3, w4 = weights
print(w1,w2,w3,w4)
```

```python
b = w1 * a
c = w2 * a
d = w3 * b + w4 * c
L = (10 - d)
```

```python
L.backward()
```

```python
for index, weight in enumerate(weights, start=1):
    gradient, *_ = weight.grad
    # gradient = weight.grad
    print(f"Gradient of w{index} w.r.t to L: {gradient}")
```

# 프로그래밍 문제 1: 위 예제코드를 참고해서 자유롭게 자신만의 여러 개의 tensor 변수 (예: w5,w6,w7,w8) 를 생성 후 이 변수들을 이용하여 자신만의 임의의 연산 (예: e, f, g, h) 과 에러 함수 (예: L) 를  만드시오. (4 points)
## 변수나 연산의 갯수는 제약이 없습니다.

```python
import torch
```

```python
weights = [Variable(FloatTensor([i]), requires_grad=True) for i in (6, 1, 4, 2)]

a = Variable(FloatTensor([4]))

w5, w6, w7, w8 = weights

e = w5 * a
f = w6 * a
g = w7 * e + w8 * f

L = 10 - g

L.backward()
```

# 프로그래밍 문제 2: 위 예제코드를 참고해서 PyTorch 에서 제공하는 Autograd 기능을 이용하여 에러 함수를 각각의 변수에 대해 미분한 값 (gradient)을 구하시오. (3 points)

```python
"""
문제 2. 이 셀 안에 코드를 작성하시오.
"""
for index, weight in enumerate(weights, start=5):
    gradient, *_ = weight.grad
    # gradient = weight.grad
    print(f"Gradient of w{index} w.r.t to L: {gradient}")
```
