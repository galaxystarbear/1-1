```python
import torch
```

```python
def validation(tensor1, tensor2):
    if None in [tensor1, tensor2]:
        print('\033[31m' + "Variable with None value was entered")
        return
    elif list in [type(tensor1), type(tensor2)]:
        tensor1 = torch.Tensor(tensor1)
        tensor2 = torch.Tensor(tensor2)
    elif tensor1.shape != tensor2.shape:
        print('\033[31m' + "Different shape", tensor1.shape, tensor2.shape)
        return

    if torch.all(torch.isclose(tensor1, tensor2)):
        print('\033[94m' + "Two tensors are identical")
    else:
        print('\033[31m' + "Something wrong..")
```

## Problem 1-1 (2 points): Einstein Summation을 이용하여 torch.matmul과 같은 계산결과를 반환하는 코드를 작성하시오.

```python
A = torch.arange(32.).reshape(8, 4)
B = torch.arange(16.).reshape(4, 4)
print('mat1 :\n', A)
print('mat2 :\n', B)
```

```python
C_answer = torch.matmul(A, B)
print('Multiplication result :\n',C_answer)
```

```python
# your code here
C_einsum =torch.einsum("ij,jk->ik",A,B)
```

```python
print("Einsum Matmul result:\n",C_einsum)
validation(C_answer, C_einsum)
```

# Problem 2 (6 points) : 각 문항별 수식에 해당하는 계산을 einsum을 이용하여 수행하시오

 ## Problem 2-1 (3 points)

\begin{align}
    A = [a_{ijk}] \in \mathbb{R}^{I\times J\times K}이고\newline
    B = [b_{jil}] \in \mathbb{R}^{J\times l \times L}일때\newline
    C = [c_{k, l}] \in \mathbb{R}^{K \times L},\newline
    c_{k, l} = \displaystyle\sum_{i = 0}^{I-1} \displaystyle\sum_{j = 0}^{J-1}{a_{i,j,k} \times b_{j, i, l}}\newline
\end{align}
을 만족하는 행렬 C의 계산을 einsum을 이용하여 수행하시오

```python
A = torch.arange(60.).reshape(3, 4, 5)
B = torch.arange(24.).reshape(4, 3, 2)
print(A)
print(B)
```

```python
C_answer = torch.sum(torch.bmm(A.permute(1, 2, 0), B), axis=0)
print(C_answer)
```

```python
#your code here
C_einsum =torch.torch.einsum('ijk,jil->kl',A,B)
```

```python
print(C_einsum)
validation(C_answer, C_einsum)
```

## Problem 2-2 (3 points)

\begin{align}
    A = [a_{bik}] \in \mathbb{R}^{B\times I\times K}이고\newline
    B = [b_{bkj}] \in \mathbb{R}^{B\times K\times J} 일 때\newline
    C = [c_{bij}] \in \mathbb{R}^{B\times I\times J},\newline
    c_{bij} = \displaystyle\sum_{k = 0}^{K-1}{a_{bik}}{b_{bkj}}
\end{align}

```python
A = torch.arange(800.).reshape(32, 5, 5)/800.0
B = torch.arange(320.).reshape(32, 5, 2)/320.0
print(A)
print(B)
```

```python
C_answer = torch.bmm(A, B)
print(C_answer)
```

```python
#your code here
C_einsum =torch.einsum('bik,bkj->bij',A,B)
```

```python
print(C_einsum)
validation(C_answer, C_einsum)
```
